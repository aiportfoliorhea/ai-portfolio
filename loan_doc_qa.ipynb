```python
import anthropic
from google.colab import userdata
api_key = userdata.get('anthropic_key')
client = anthropic.Anthropic(api_key=api_key)
# Hardcoded loan document text
loan_doc = """
Loan Agreement Summary:
- Borrower: Jane Doe
- Loan Amount: $250,000
- Interest Rate: 6.5% fixed
- Term: 30 years
- Monthly Payment: $1,580
- Prepayment Penalty: None
- Late Fee: 5% of payment if more than 15 days late
"""

question = "What is the monthly payment and is there a prepayment penalty?"
message = client.messages.create(
    model="claude-sonnet-4-5",
    max_tokens=100,
    messages=[
        {"role": "user", "content": f"Here is a loan document:\n\n{loan_doc}\n\nQuestion: {question}"}
    ]
)

print(message.content[0].text)
print(f"Tokens used - output: {message.usage.output_tokens}")
```

```python
questions = [
    "What is the monthly payment?",
    "Is there a prepayment penalty?",
    "What happens if I pay late?"
]

for q in questions:
    message = client.messages.create(
        model="claude-sonnet-4-5",
        max_tokens=200,
        messages=[
            {"role": "user", "content": f"Here is a loan document:\n\n{loan_doc}\n\nQuestion: {q}"}
        ]
    )
    print(f"Q: {q}")
    print(f"A: {message.content[0].text}\n")
    print(f"Tokens used - output: {message.usage.output_tokens}")
```
